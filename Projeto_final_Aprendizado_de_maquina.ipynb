## Criterion

Como cada vez que realizavamos o modelo os valores diferiam e mostravam hora melhor em um, hora melhor outro parametro, fizemos uma iteração para encontrarmos o melhor hiperparametro para esse caso e após inúmeras repetições encontramos que o melhor o criterion foi o padrao , 'gini'. 

- MEDIA TOTAL HIPERPARAMETRO GINI :  0.637
- MEDIA TOTAL HIPERPARAMETRO ENTROPY :  0.631
- MEDIA TOTAL HIPERPARAMETRO LOG_LOSS :  0.629

Apesar de pouca diferença aos longos do testes o hiperparametro gini continuou apresentando os melhores resultados

```python
class HyperparamCriterion:
  sum_hyperparam_gini = 0
  sum_hyperparam_entropy = 0
  sum_hyperparam_log_loss = 0

  def __init__(self, criterion_options):
    self.criterion_options = criterion_options

  def criterion_hyperparam_test(self, criterion, iterator1):
    modelo_cart_criterion = DecisionTreeClassifier(criterion=criterion)
    cv_results_cart_criterion = cross_val_score(modelo_cart_criterion, variaveis_train, comprou_bike_train, cv=kfold, scoring=scoring)
    mean_cv_results = cv_results_cart_criterion.mean()
    standard_deviation = cv_results_cart_criterion.std()

    
    if criterion == 'gini':
      self.sum_hyperparam_gini = self.sum_hyperparam_gini + mean_cv_results   
    if criterion == 'entropy':
      self.sum_hyperparam_entropy = self.sum_hyperparam_entropy + mean_cv_results 
    if criterion == 'log_loss':
      self.sum_hyperparam_log_loss = self.sum_hyperparam_log_loss + mean_cv_results 
   
    print("Iteração %i . hiperparametro criterion: %s  : Acurácia: %f , Desvio padrão: %f"%(iterator1, criterion, mean_cv_results, standard_deviation))

  def compare_hyperparams_criterion(self, iterations):
    for iterator1 in range(iterations):
      for criterion in criterion_options:
        acuracy = self.criterion_hyperparam_test(criterion, iterator1)
    print("MEDIA TOTAL HIPERPARAMETRO GINI : ",self.sum_hyperparam_gini/iterations)
    print("MEDIA TOTAL HIPERPARAMETRO ENTROPY : ",self.sum_hyperparam_entropy/iterations)    
    print("MEDIA TOTAL HIPERPARAMETRO LOG_LOSS : ",self.sum_hyperparam_log_loss/iterations)    

criterion_options = ['gini','entropy','log_loss']
testCriterion = HyperparamCriterion(criterion_options)
testCriterion.compare_hyperparams_criterion(100)
```

```python
  

class HyperparamSplitter:
  sum_hyperparam_best = 0
  sum_hyperparam_random = 0

  def __init__(self, splitter_options):
    self.splitter_options = splitter_options

  def splitter_hyperparam_test(self,splitter,iterator2):

      modelo_cart_splitter = DecisionTreeClassifier(splitter=splitter)
      cv_results_cart_splitter = cross_val_score(modelo_cart_splitter, variaveis_train, comprou_bike_train, cv=kfold, scoring=scoring)
      
      mean_cv_results = cv_results_cart_splitter.mean();
      standard_deviation = cv_results_cart_splitter.std()

      if splitter == 'best':
        self.sum_hyperparam_best = self.sum_hyperparam_best + mean_cv_results
      if splitter == 'random':
        self.sum_hyperparam_random = self.sum_hyperparam_random +mean_cv_results 
      
      print("Iteração %i . hiperparametro splitter: %s  : Acurácia: %f , Desvio padrão: %f"%(iterator2, splitter, mean_cv_results, standard_deviation))

  def compare_splitters(self,iterations):
    for iterator2 in range(iterations):
        for splitter in self.splitter_options:
            self.splitter_hyperparam_test(splitter,iterator2)

    print("MEDIA TOTAL HIPERPARAMETRO BEST : ",self.sum_hyperparam_best/iterations)
    print("MEDIA TOTAL HIPERPARAMETRO RANDOM : ",self.sum_hyperparam_random/iterations)

splitter_options = ['best','random']

teste1 = HyperparamSplitter(splitter_options)

teste1.compare_splitters(100)
```

## Max Depth

Para o hiperparametro maxdepth, fizemos uma serie de repeticoes que serao as linhas da matriz e cada numero de decisoes da arvore a ser testado serão as colunas, dessa forma podemos tirar a media e ver qual delas se saem melhor quando repetida 10 vezes, os testes variaram de 1 a 64. Obtivemos que o hiperparametro max_depth quando variado entre os inteiros tem alta latencia, enquanto o valor padrão None mantém uma regularidade maior
- max_depth:None.  Acurácia :  0.64
- Demais no array impresso 

```python
class HyperparamMaxDepth:
  #matriz para acumulo de testes, as linhas representam as repeticoes de cada hyperparametro max_depth e as colunas o numero do max_depth propriamente dito
  
  
  def __init__(self,number_of_decisions, times_to_repeat):
      self.times_to_repeat = times_to_repeat
      self.number_of_decisions = number_of_decisions
      self.matrix = [[None for j in range(number_of_decisions)] for i in range(times_to_repeat)]
      
      #array com todas as medias
      self.arrayMeansMaxDepth = [0 for h in range(number_of_decisions)]


  def max_depth_hyperparam_test(self, decisions, row):
    modelo_cart_max_depth = DecisionTreeClassifier(max_depth=decisions+1)
    cv_results_cart_max_depth = cross_val_score(modelo_cart_max_depth, variaveis_train, comprou_bike_train, cv=kfold, scoring=scoring)
    
    mean_cv_results = cv_results_cart_max_depth.mean()
    standard_deviation_cv = cv_results_cart_max_depth.std()

    self.matrix[row][decisions] = mean_cv_results

  def max_depth_default(self):
    modelo_cart_max_depth = DecisionTreeClassifier(max_depth=None)
    cv_results_cart_max_depth = cross_val_score(modelo_cart_max_depth, variaveis_train, comprou_bike_train, cv=kfold, scoring=scoring)
    
    mean_cv_results = cv_results_cart_max_depth.mean()
    standard_deviation_cv = cv_results_cart_max_depth.std()

    print("max_depth:None,  Acurácia : ", mean_cv_results)

  #Criação da matriz 
  def generateDecisionsMatrix(self):
    for i in range(self.times_to_repeat):
      for decisions in range(self.number_of_decisions):
        self.max_depth_hyperparam_test(decisions,i)
      print(i)  
      decisions =0  

  def allResultsMeans(self):
    for i in range(len(self.matrix)):
      for j in range(len(self.matrix[i])):
        self.arrayMeansMaxDepth[j] = self.arrayMeansMaxDepth[j] + self.matrix[i][j]


    for k in range(len(self.arrayMeansMaxDepth)):
      self.arrayMeansMaxDepth[k] = self.arrayMeansMaxDepth[k]/self.times_to_repeat  
    print("Média das iterações para os max_dephs de 1 a 64",self.arrayMeansMaxDepth)
    return self.arrayMeansMaxDepth

#Algoritmo com 10 repetiçoes max def indo de 1 a 64
testeMaxDepth = HyperparamMaxDepth(64,10)
testeMaxDepth.generateDecisionsMatrix()
arrayMeansMaxDepth = testeMaxDepth.allResultsMeans()

testeMaxDepth.max_depth_default()
```

```python
class HyperparamMaxLeafNodes:
  #matriz para acumulo de testes, as linhas representam as repeticoes de cada hyperparametro max_depth e as colunas o numero do max_depth propriamente dito
  
  
  def __init__(self,number_of_max_leaf_nodes, times_to_repeat):
      self.times_to_repeat = times_to_repeat
      self.number_of_max_leaf_nodes = number_of_max_leaf_nodes
      self.matrix3 = [[None for j in range(number_of_max_leaf_nodes)] for i in range(times_to_repeat)]
      
      #array com todas as medias
      self.arrayMeans3 = [0 for h in range(number_of_max_leaf_nodes)]


  def max_leaf_nodes_hyperparam_test(self, max_leaf_nodes, row):
    modelo_cart_max_leaf_nodes = DecisionTreeClassifier(min_samples_split=2,min_samples_leaf=1,max_depth=28,splitter="random", max_leaf_nodes=max_leaf_nodes+2)
    cv_results_cart_max_leaf_nodes = cross_val_score(modelo_cart_max_leaf_nodes, variaveis_train, comprou_bike_train, cv=kfold, scoring=scoring)
    
    mean_cv_results = cv_results_cart_max_leaf_nodes.mean()
    standard_deviation_cv = cv_results_cart_max_leaf_nodes.std()

    self.matrix3[row][max_leaf_nodes] = mean_cv_results

  #Criação da matriz 
  def generateDecisionsMatrix(self):
    for j in range(self.times_to_repeat):
      for max_leaf_nodes in range(self.number_of_max_leaf_nodes):
        self.max_leaf_nodes_hyperparam_test(max_leaf_nodes,j)
      print(j)  
      decisions =0  

  def allResultsMeans(self):
    for i in range(len(self.matrix3)):
      for j in range(len(self.matrix3[i])):
        self.arrayMeans3[j] = self.arrayMeans3[j] + self.matrix3[i][j]


    for k in range(len(self.arrayMeans3)):
      self.arrayMeans3[k] = self.arrayMeans3[k]/self.times_to_repeat
    print(self.arrayMeans3)
    return self.arrayMeans3
testeMaxLeafNodes = HyperparamMaxLeafNodes(51,20)
testeMaxLeafNodes.generateDecisionsMatrix()
arrayMaxLeafNodes = testeMaxLeafNodes.allResultsMeans()
```

## Min impurity decrease

Para o min_impurity_decrease, temos duas opções o valor calculado ou o padrão, como veremos em seguida o valor padrão apresenta maior acurácia

```python


clf = DecisionTreeClassifier(min_impurity_decrease=0.01)
clf.fit(variaveis_test, comprou_bike_test)

# impureza da raiz da árvore de decisão
root_impurity = clf.tree_.impurity[0]

# impureza dos nós filhos esquerdo e direito da raiz da árvore de decisão
left_impurity = clf.tree_.impurity[clf.tree_.children_left[0]]
right_impurity = clf.tree_.impurity[clf.tree_.children_right[0]]

# número de amostras nos nós filhos esquerdo e direito da raiz da árvore de decisão
n_left = clf.tree_.n_node_samples[clf.tree_.children_left[0]]
n_right = clf.tree_.n_node_samples[clf.tree_.children_right[0]]

# Fórmula do min_impurity_decrease encontrada no sklearn doc
weighted_average_impurity = (n_left * left_impurity + n_right * right_impurity) / (n_left + n_right)
weighted_samples = n_left + n_right

min_impurity_decrease = (root_impurity - weighted_average_impurity) / weighted_samples

print("Valor de min_impurity_decrease calculado é:", min_impurity_decrease)

```

```python
class HyperparamCcpAlpha:
  sum_hyperparam_0 = 0
  sum_hyperparam_1 = 0
  sum_hyperparam_2 = 0
  sum_hyperparam_3 = 0
  sum_hyperparam_4 = 0
  sum_hyperparam_5 = 0
  sum_hyperparam_6 = 0
  sum_hyperparam_7 = 0
  sum_hyperparam_8 = 0
  def __init__(self, ccp_alpha_options):
    self.ccp_alpha_options = ccp_alpha_options

  def ccp_alpha_options_hyperparam_test(self,ccp_alpha):

      modelo_cart_ccp_alpha = DecisionTreeClassifier(min_samples_split=2,max_depth=28,splitter="random",random_state=35,ccp_alpha=ccp_alpha)
      cv_results_cart_ccp_alpha = cross_val_score(modelo_cart_ccp_alpha, variaveis_train, comprou_bike_train, cv=kfold, scoring=scoring)
      
      mean_cv_results = cv_results_cart_ccp_alpha.mean();
      standard_deviation = cv_results_cart_ccp_alpha.std()

      if ccp_alpha == 0.0:
        self.sum_hyperparam_0 =  mean_cv_results
      if ccp_alpha == 0.0001:
        self.sum_hyperparam_1 =  mean_cv_results
      if ccp_alpha == 0.001:
        self.sum_hyperparam_2 = mean_cv_results 
      if ccp_alpha == 0.01:
        self.sum_hyperparam_3 = mean_cv_results 
      if ccp_alpha ==  0.1:
        self.sum_hyperparam_4 = mean_cv_results 
      if ccp_alpha ==  1:
        self.sum_hyperparam_5 = mean_cv_results 
      if ccp_alpha ==  10:
        self.sum_hyperparam_6 = mean_cv_results
      if ccp_alpha ==  100:
        self.sum_hyperparam_7 = mean_cv_results 
      if ccp_alpha ==  1000:
        self.sum_hyperparam_8 = mean_cv_results   

      
      print("hiperparametro ccp_alpha: %s  : Acurácia: %f , Desvio padrão: %f"%(ccp_alpha, mean_cv_results, standard_deviation))

  def compare_ccp_alpha(self):
    
    for ccp_alpha in self.ccp_alpha_options:
      self.ccp_alpha_options_hyperparam_test(ccp_alpha)
    print("MEDIA TOTAL HIPERPARAMETRO CCP_ALPHA = 0.0 : ",self.sum_hyperparam_0)

    print("MEDIA TOTAL HIPERPARAMETRO CCP_ALPHA = 0.0001 : ",self.sum_hyperparam_1)
    print("MEDIA TOTAL HIPERPARAMETRO CCP_ALPHA = 0.001 : ",self.sum_hyperparam_2)
    print("MEDIA TOTAL HIPERPARAMETRO CCP_ALPHA = 0.01 : ",self.sum_hyperparam_3)

    print("MEDIA TOTAL HIPERPARAMETRO CCP_ALPHA = 0.1 : ",self.sum_hyperparam_4)
    print("MEDIA TOTAL HIPERPARAMETRO CCP_ALPHA = 1 : ",self.sum_hyperparam_5)
    print("MEDIA TOTAL HIPERPARAMETRO CCP_ALPHA = 10 : ",self.sum_hyperparam_6)
    print("MEDIA TOTAL HIPERPARAMETRO CCP_ALPHA = 100 : ",self.sum_hyperparam_7)
    print("MEDIA TOTAL HIPERPARAMETRO CCP_ALPHA = 1000 : ",self.sum_hyperparam_8)



ccp_alpha_options =[0.0,0.0001,0.001,0.01,0.1,1,10,100,1000]
test_ccp_alpha = HyperparamCcpAlpha(ccp_alpha_options)    
test_ccp_alpha.compare_ccp_alpha()
```
